## *Pandas*:

#### Data Series: 

* It is having only 1 ROW or only having 1 COLOUMN  

* It can have the data in any type like ( Int , FLoat , string , complex etc ) 

* Data Series is always associated with the index and if we did'nt provide the index it will start by default 0 

* In data Series we can give the index of various types(int, string , datatime etc) 

    * To create the series we used the function : **pd.Series() -- For creating the Series**

**Dataframe** : 
- it is a two dimensional array 
- it will contains rows and coloumns 
- If there is more than one row or more than one coloumn then it will called as Data Frame 

- For creating the data frame we have method : **Pd.DataFrame()** for creating dataframe

```python
import pandas as pd

# Creating a DataFrame 
data = {
    "Date": ["2025-03-25", "2025-03-26", "2025-03-27", "2025-03-28", "2025-03-29"],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
    "Temperature": [15, 22, 10, 28, 30],  # in Celsius
    "Windspeed": [10, 8, 12, 15, 9],  # in km/h
    "Event": ["Sunny", "Cloudy", "Rain", "Thunderstorm", "Sunny"]
}

df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd

# List of tuples (each tuple is a row)
data = [
    ("2025-03-25", "New York", 15, 10, "Sunny"),
    ("2025-03-26", "Los Angeles", 22, 8, "Cloudy"),
    ("2025-03-27", "Chicago", 10, 12, "Rain"),
    ("2025-03-28", "Houston", 28, 15, "Thunderstorm"),
    ("2025-03-29", "Miami", 30, 9, "Sunny")
]

# Creating DataFrame with column names
df = pd.DataFrame(data, columns=["Date", "City", "Temperature", "Windspeed", "Event"])

```
